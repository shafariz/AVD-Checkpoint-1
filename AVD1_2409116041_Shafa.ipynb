# **Business Understanding**


**Business Objective**

  Tujuan utama proyek ini adalah memahami tren dan faktor-faktor yang berkontribusi terhadap popularitas drama Korea. Analisis ini dapat digunakan untuk membantu industri hiburan dalam mengembangkan strategi produksi, pemasaran, dan distribusi yang lebih efektif. Selain itu, proyek ini bertujuan untuk mengidentifikasi pola dalam genre, rating, jumlah episode, dan faktor lain yang mempengaruhi penerimaan penonton.


**Asses Situation**

Industri hiburan Korea semakin berkembang dengan persaingan yang ketat, baik di tingkat domestik maupun internasional. Platform streaming global seperti Netflix, Disney+, dan Viu semakin meningkatkan aksesibilitas KDrama ke pasar yang lebih luas. Dengan adanya perubahan preferensi penonton dan kemajuan teknologi, perusahaan produksi dan distributor perlu memahami tren yang sedang berkembang untuk mengoptimalkan strategi konten mereka. Oleh karena itu, analisis data menjadi penting dalam mengidentifikasi faktor yang berpengaruh terhadap kesuksesan drama Korea.



**Data Mining Goals**

Proyek ini bertujuan untuk mengidentifikasi faktor utama yang berkontribusi terhadap rating tinggi suatu drama Korea dengan menganalisis hubungan antara genre, jumlah episode, durasi, dan elemen lainnya. Selain itu, proyek ini juga mengeksplorasi tren produksi berdasarkan tahun rilis dan jaringan penyiaran, serta melihat pola popularitas berdasarkan pemeran, sutradara, dan penulis naskah. Dengan memahami pola-pola ini, industri hiburan dapat mengoptimalkan strategi produksi dan pemasaran mereka agar lebih sesuai dengan preferensi penonton.


**Project Plan**

Langkah pertama dalam proyek ini adalah melakukan pembersihan dan persiapan data untuk memastikan kualitas dan konsistensi informasi yang dianalisis. Setelah itu, dilakukan eksplorasi data untuk memahami distribusi rating, tren genre, serta hubungan antara berbagai faktor yang mempengaruhi popularitas KDrama. Selanjutnya, teknik data mining seperti clustering atau regresi akan diterapkan untuk menemukan pola dalam data yang dapat memberikan wawasan bagi industri hiburan. Hasil dari analisis ini kemudian akan diinterpretasikan dan disusun dalam laporan yang dapat digunakan oleh perusahaan produksi dan platform streaming dalam menyusun strategi pengembangan dan pemasaran drama Korea.
